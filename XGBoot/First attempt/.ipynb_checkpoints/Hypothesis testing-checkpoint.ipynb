```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
folder_19_20 = "/Users/calumthompson/Documents/Fantasy_football/GitHub_data/data/2019-20/"
folder_18_19 = "/Users/calumthompson/Documents/Fantasy_football/GitHub_data/data/2018-19/"
```

# Is there a linear relationship between performance between seasons?

Not really

```python
pd.read_csv(folder_19_20 + "cleaned_players.csv").sort_values('total_points', ascending = False).head()
```

```python
season_summary_19_20 = pd.read_csv(folder_19_20 + "cleaned_players.csv")

season_summary_18_19 = pd.read_csv(folder_18_19 + "cleaned_players.csv", usecols = ['first_name','second_name','total_points'])

season_summary_19_20 = season_summary_19_20.rename(columns={'total_points' : 'total_points_19_20'})
season_summary_18_19 = season_summary_18_19.rename(columns={'total_points' : 'total_points_18_19'})
                                     

season_total_comparison_df = pd.merge(season_summary_19_20,season_summary_18_19, on = ['first_name', 'second_name'])
```

```python
y = season_total_comparison_df['total_points_19_20'].values.reshape(-1,1)
x = season_total_comparison_df['total_points_18_19'].values.reshape(-1,1)
regressor = LinearRegression() 

plt.scatter(x,y )
plt.xlabel("total_points_18_19")
plt.ylabel("total_points_19_20")

regressor.fit(x,y)
    
fit = regressor.predict(x) 

slope = regressor.coef_[0][0]
error = metrics.mean_absolute_error(y, fit)  
r_squared = metrics.r2_score(y,fit)

plt.plot (x, fit, c = 'k', label = 'linear fit')

print('slope = ',slope)
print('error = ',error)
print('r_squared = ',r_squared)
```

## What if we take out the shit munchers?

```python
season_total_comparison_top_half_df = season_total_comparison_df.sort_values('total_points_18_19',ascending = False)
n = len(season_total_comparison_top_half_df)
season_total_comparison_top_half_df = season_total_comparison_top_half_df.head(int(n/3))
season_total_comparison_top_half_df.total_points_18_19.mean()
```

```python
y = season_total_comparison_top_half_df['total_points_19_20'].values.reshape(-1,1)
x = season_total_comparison_top_half_df['total_points_18_19'].values.reshape(-1,1)
regressor = LinearRegression() 

plt.scatter(x,y )
plt.xlabel("total_points_18_19")
plt.ylabel("total_points_19_20")

regressor.fit(x,y)
    
fit = regressor.predict(x) 

slope = regressor.coef_[0][0]
error = metrics.mean_absolute_error(y, fit)  
r_squared = metrics.r2_score(y,fit)

plt.plot (x, fit, c = 'k', label = 'linear fit')

print('slope = ',slope)
print('error = ',error)
print('r_squared = ',r_squared)
```

# How much does cost reflect last year's points?

```python
y = season_total_comparison_df['total_points_19_20'].values.reshape(-1,1)
x = season_total_comparison_df['now_cost'].values.reshape(-1,1)
regressor = LinearRegression() 

plt.scatter(x,y )
plt.xlabel("now_cost")
plt.ylabel("total_points_19_20")

regressor.fit(x,y)
    
fit = regressor.predict(x) 

slope = regressor.coef_[0][0]
error = metrics.mean_absolute_error(y, fit)  
r_squared = metrics.r2_score(y,fit)

plt.plot (x, fit, c = 'k', label = 'linear fit')

print('slope = ',slope)
print('error = ',error)
print('r_squared = ',r_squared)
```

```python
best_players = gw_data.groupby('name')['total_points'].agg({np.sum, np.std}).sort_values('sum', ascending = False).head(30).index.values

gw_data.groupby('name')['total_points'].agg({np.sum, np.std}).sort_values('sum', ascending = False).head(10)
```
